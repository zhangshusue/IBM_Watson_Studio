# Data Science Methodology 
This is the lecture notes from [Dr. John Rollins](https://www.slideshare.net/slideshow/embed_code/key/GO21lKl4AdYptk) on  "*Data Science Mathodology*". 

## The Data Science Methodology aims to answer the following 10 questions: 

### From Problem to Approach: 
1. What is the problem you are trying to solve? 
2. How can you use data to answer the questions? 

### Working with Data: 
3. What data do you need to answer the questions? 
4. Where is the data coming from (identify all sources) and how will you get it? 
5. Is the data that you collected representative of the problem to be solved? 
6. What additional work is required to manipulate and work with the data? 

### Deriving the Answer:
7. In what way can the data be visualized to get to the answer that is required? 
8. Does the model used really answer the initial question or does it need to be adjusted? 
9. Can you put the model into practice?
10. Can you get constructive feedback into answering the question?


## Five major stages/processes of a standard Data Science project:
1. **From Problem to Approach**:
  * Business Understanding: the most important stage, understand the question/need
  * Analytical Approach: models can be determined as the question is fully understood (predictive, descriptive, classification models and machine learning, etc.)
  
2. **From Requirements to Collection**: 
  * Data Requirements: identify the necessary data content, formats and sources of initial data
  * Data Collection: identify potential data sources and review/revise data requirement
  
3. **From Understanding to Preparation**:
  * Data Understanding: understand the data: is the data collective representative for the questions? (descriptive statistics, missing/misleading/invalid values)  
  * Data Preparation: cleansing data (50%-90% of time) and transforming data (feature engineering/domain knowledge), pay attention to the details  
  
4. **From Modeling to Evaluation**:
  * Modeling: to build descriptive or predictive model, use training set for predictive modeling, constant refinement/adjustments/tweaking (need to iterate between this step and data preparation to balance between accuracy/sensitivity/specificity)  
  * Evaluation: to assess the model quality and to see if it meets the initial request. Evaluation phases:
    * diagnostic measures phase: to ensure the model is working as intended
    * statistical significance testing: false-positive (type I error), false-negative (type II error), ROC curve (receiver operating characteristic curve) 
  
5. **From Delpoyment to Feedback**: 
  * Delpoyment: put into testing environment after initial evaluation (need help from different stakeholders) 
  * Feedback: collect feedback from users to refine model and re-assess ("The more you know, the more you want to know" - John Rolllins ðŸ˜Š)
    * "Top-down" vs. "bottom-up" approach
    
  
 ### Summary: The Data Science Methodology process is highly iterative and requires constant inputs from different stakeholders.
